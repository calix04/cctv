```python
import pandas as pd
import numpy as np
import folium
import plotly.express as px
import os
import glob
import json
```

```python
df = pd.read_csv("서울특별시 (안심이) CCTV 설치 현황.csv", encoding = 'cp949')
df
```

```python
dataset = df[[
    '자치구','안심 주소','CCTV 용도','위도', '경도','CCTV 수량'
]] # dataset에 자치구 , 안심주소, cctv용도, 위도, 경도 , cctv 수량 컬럼만 입력
dataset.head()
```

```python
dataset.index = range(len(dataset))
print('cctv 수 :', len(dataset))
dataset
```

```python
cctv_gu = dataset[(dataset['자치구'] == '중랑구')] #구의 cctv수
cctv_gu.index = range(len(cctv_gu))
print('중랑구 cctv 수 :', len(cctv_gu))
cctv_gu.head()
```

```python
df_cctv = dataset[(dataset['CCTV 수량']>=int(1))]
df_cctv.index = range(len(df_cctv))
print('CCTV 수량 :', len(df_cctv))
df_cctv.head()
```

```python
df_seoul_cctv = dataset[dataset['자치구'].str.contains('구')]
df_seoul_cctv.index = range(len(df_seoul_cctv)) #인덱스
df_seoul_cctv.head()
```

```python
cctv_region = df_cctv.groupby('자치구')['CCTV 수량'].count().to_frame().sort_values(by='자치구', ascending=False)
cctv_region = cctv_region.reset_index()
cctv_region = cctv_region.set_index('자치구')
cctv_region
```

```python
cctv_region
```

```python
geo_path = '02. skorea_municipalities_geo_simple.json' 
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
loc = [37.498095, 127.027610] # 위도(N), 경도(E)
# data_size = len(df_seoul_cctv)

map_cctv = folium.Map(location=loc, # 좌표 대입
                            zoom_start=10) # 얼만큼의 확대로 시작할건지를 설정

map_cctv.choropleth(geo_data=geo_str,
              data = cctv_region['CCTV 수량'],
              columns=[cctv_region.index, cctv_region['CCTV 수량']],
              fill_color='PuRd',
              key_on='feature.id')

# #마커 설정
# for i in range(data_size):

#     folium.Marker(list(df_cctv.iloc[i][['위도', '경도']]), # 엑셀파일의 위도, 경도 위치에 맞게 마커를 찍어주도록 설정
#                  popup = df_cctv.iloc[i][['안심 주소']], # 학교 명이 뜨도록 설정
#                  icon = folium.Icon(color = 'black')).add_to(map_cctv) # 색상을 블랙으로 설정

map_cctv
map_cctv.save('cctv.html')
```
